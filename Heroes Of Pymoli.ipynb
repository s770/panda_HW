```python
# Purchasing Analysis (Gender)
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
```

```python
gender_purchase_count = pymoli_df.groupby(["Gender"]).count()["Price"]
gender_purchase_average = pymoli_df.groupby(["Gender"]).mean()["Price"]
gender_total = pymoli_df.groupby(["Gender"]).sum()["Price"]
gender_avg = gender_total/gender_count

gender_purchase_summary = pd.DataFrame({"Purchase Count": gender_purchase_count,
                                       "Average Purchase Price":gender_purchase_average,
                                       "Total Purchase Value":gender_total,
                                       "Avg Total Purchase per Person":gender_avg})
gender_purchase_summary["Average Purchase Price"] = gender_purchase_summary["Average Purchase Price"].map("${:,.2f}".format)
gender_purchase_summary["Total Purchase Value"] = gender_purchase_summary["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_summary["Avg Total Purchase per Person"] = gender_purchase_summary["Avg Total Purchase per Person"].map("${:,.2f}".format)

gender_purchase_summary
```

```python
# Age Demographics
# Establish bins for ages
# # Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table
```

```python
pymoli_df.head()
```

```python
pymoli_df["Age"].max(), pymoli_df["Age"].min()
```

```python
bins = [0,9, 14, 19,24,29,34,39,200]
age_group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
pymoli_df["Ages"] = pd.cut(pymoli_df["Age"], bins, labels = age_group)
pymoli_df.head()
```

```python
age_bin_count = pymoli_df.groupby(["Ages"]).count()["Screen Name"]

age_bin_percent = (age_bin_count / 576) * 100

age_summary = pd.DataFrame({"Total Count":age_bin_count,
                             "Percentage of Players":age_bin_percent}).round(2)
age_summary
```

```python
# Purchasing Analysis (Age)
# Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data fram
```

```python
age_purchase_count = pymoli_df.groupby(["Ages"]).count()["Price"]

age_purchase_avg = pymoli_df.groupby(["Ages"]).mean()["Price"]

age_purchase_total = pymoli_df.groupby(["Ages"]).sum()["Price"]

avg_total_per_person = age_purchase_total/age_purchase_count

purchase_summary = pd.DataFrame({"Purchase Count": age_purchase_count,
                                "Average Purchase Price": age_purchase_avg,
                                "Total Purchase Value": age_purchase_total,
                                "Avg Total Purchase per Person":avg_total_per_person})

purchase_summary["Average Purchase Price"] = purchase_summary["Average Purchase Price"].map("${:,.2f}".format)
purchase_summary["Total Purchase Value"] = purchase_summary["Total Purchase Value"].map("${:,.2f}".format)
purchase_summary["Avg Total Purchase per Person"] = purchase_summary["Avg Total Purchase per Person"].map("${:,.2f}".format)


purchase_summary
```

```python
# Top Spenders
# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# # Display a preview of the summary data frame

sn_count = pymoli_df.groupby(["Screen Name"]).count()["Price"]
sn_avg = pymoli_df.groupby(["Screen Name"]).mean()["Price"]
sn_total = pymoli_df.groupby(["Screen Name"]).sum()["Price"]

top_spenders_summary = pd.DataFrame({"Purchase Count": sn_count,
                                     "Average Purchase Price":sn_avg,
                                     "Total Purchase Value":sn_total})
top_spenders_summary["Average Purchase Price"] = top_spenders_summary["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_summary["Total Purchase Value"] = top_spenders_summary["Total Purchase Value"].map("${:,.2f}".format)

top_spenders_summary_sort = top_spenders_summary.sort_values(["Purchase Count"], ascending=False)
top_spenders_summary_sort.head()
```

```python
#Most Popular Items
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
# Create a summary data frame to hold the results
# Sort the purchase count column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame

purchase_count = pymoli_df.groupby(["Item ID", "Item Name"]).count()["Price"]
total_purchase_value = pymoli_df.groupby(["Item ID", "Item Name"]).sum()["Price"]
item_price = pymoli_df.groupby(["Item ID", "Item Name"]).mean()["Price"]

purchase_count.head()
popular_summary = pd.DataFrame({"Purchase Count":purchase_count,
                                "Item Price":item_price,
                                "Total Purchase Value":total_purchase_value})
popular_summary["Total Purchase Value"] = popular_summary["Total Purchase Value"].map("${:,.2f}".format)
popular_summary["Item Price"] = popular_summary["Item Price"].map("${:,.2f}".format)

popular_summary_sort = popular_summary.sort_values(["Purchase Count"], ascending=False)
popular_summary_sort.head()
```

```python
# Most Profitable Items
# Sort the above table by total purchase value in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the data frame

purchase_count = pymoli_df.groupby(["Item ID", "Item Name"]).count()["Price"]
total_purchase_value = pymoli_df.groupby(["Item ID", "Item Name"]).sum()["Price"]


popular_summary = pd.DataFrame({"Purchase Count":purchase_count,
                                "Item Price":item_price,
                               "Total Purchase Value":total_purchase_value})
popular_summary["Total Purchase Value"] = popular_summary["Total Purchase Value"].map("${:,.2f}".format)
popular_summary["Item Price"] = popular_summary["Item Price"].map("${:,.2f}".format)

popular_summary_sort = popular_summary.sort_values(["Purchase Count"], ascending=False)
popular_summary_sort.head()
```
